```python
from pathlib import Path
from urllib.request import urlretrieve

CORPUS_DIR = Path('/content/data')
CORPUS_DIR.mkdir(parents=True, exist_ok=True)
TEXT_PATH = CORPUS_DIR / 'tiny_shakespeare.txt'

if not TEXT_PATH.exists():
    urlretrieve('https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt', TEXT_PATH)

print(f'text_path={TEXT_PATH}')
print(f'chars={len(TEXT_PATH.read_text(encoding="utf-8")):,}')
```
